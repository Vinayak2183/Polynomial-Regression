## Import Libraries

```python
import numpy as mp 
import pandas as pd
import matplotlib.pyplot as plt
```

## Import dataset and preprocessing

```python
position=pd.read_csv("D:\\machine learning\\course\\regression\\polynomial regression\\Position_Salaries.csv")
```

```python
x=position.iloc[:,1:-1].values
y=position.iloc[:,-1].values
```

## Training the linear Regression 
###### going to compare prediction from linear regression model with polynomial regression model

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

## Training the Polynomial regression model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)
x_poly=poly_reg.fit_transform(x)
lin_reg2=LinearRegression()
lin_reg2.fit(x_poly,y)
```

## Visiualing the Best fit line of the Linear regression model
##### the line we see is very inaccurate according to the dataset hence it will give very inaccurate result

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Linear Regression')
plt.xlabel('postion level')
plt.ylabel('salary')
plt.show()
```

## Visiualing the polynomial regression model
##### the best fit line in this model is better in comparision to linear regression model hence it will give us better result

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(x_poly),color='blue')
plt.title('Polynomial regression')
plt.xlabel('Position level')
plt.ylabel('saalry')
plt.show()
```

## Training another Polynomial regression model on same data but with higher degree

```python
poly_reg2=PolynomialFeatures(degree=4)
x_poly2=poly_reg2.fit_transform(x)
lin_reg3=LinearRegression()
lin_reg3.fit(x_poly2,y)
```

## Visiualizing the better model with higer degree
##### as we can see the best fit curve is way more accurate than the other model

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg3.predict(x_poly2),color='blue')
plt.title('Polynomial regression(with more higher degree)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

## Predicting values of salary with respect to the position level using linear regression model
##### as we can see the given value as much higher than the expected value according to dataset

```python
lin_reg.predict([[6.5]])
```

## predicting value of salary with respect to the position level using Polynomial regression model
##### as we can see the predict value is more accurate to the expected value according to dataset 

```python
lin_reg3.predict(poly_reg2.fit_transform([[6.5]]))
```
